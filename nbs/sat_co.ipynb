```python
from useful_scit.imps import *
```

```python
path = '../data/co_sat/'
```

```python
files = glob.glob(pjoin(path,'*.nc'))
```

```python
files= np.sort(files)
```

```python
f = files[0]
xss = []
for f in files:
    xs = xr.open_dataset(f)
    xss.append(xs)
```

```python
xc = xr.concat(xss,dim='time')
```

```python
for xla in xc[coc].transpose('lat','lon','time'):
    for xlo in xla:
        line=xlo.plot(color=[0,0,0,.1])
        line = line[0]
line.axes.grid()
```
